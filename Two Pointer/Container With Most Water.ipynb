- You are given an integer array heights where heights[i] represents the height of the ith bar.
- You may choose any two bars to form a container. 
- Return the maximum amount of water a container can store.

```python
from typing import List
```

```python
class Solution:
    def maxArea(self, nums: List[int]) -> int:
        max_area = 0
        
        ## Intialize l and r pointers
        l, r = 0, len(nums)-1

        while l<r:
            ## Move whichever is min
            area = (r-l) * min (nums[l], nums[r])
            max_area = max(area, max_area)
            if nums[l] < nums[r]:
                l +=1
            elif nums[l] > nums[r]:
                r -=1
                
            ## If both are equal, move the one which has the greater next element 
            else:
                if nums[l+1] > nums[r-1] and  l < r:
                    l+=1
                else:
                    r -=1
        return max_area
                
        
```

```python
height = [1,7,2,5,4,7,3,6]
sol = Solution()
max_area = sol.maxArea(height)
```
